```python
import geopandas as gpd
import pandas as pd

# Cargar shapefile y mantener columnas clave
gdf = gpd.read_file("C:/Users/paola/Analisis Geoespacial/Proyecto_01/gdf_resultado_modelos_FINAL.shp")

# Asegúrate de que el CRS sea EPSG:4326 para folium
gdf = gdf.to_crs(epsg=4326)

# Agrupar por municipio
df_muni = gdf.groupby("mncp_nr").agg({
    "Rend": "mean",
    "Bj_Rndm": "mean"  # % bajo rendimiento
}).reset_index()
df_muni.columns = ["municipio", "rendimiento_prom", "porc_bajo_rend"]

# Combinar con geometría (uno por municipio)
gdf_muni = gdf.dissolve(by="mncp_nr", as_index=False).merge(df_muni, left_on="mncp_nr", right_on="municipio")
```

```python
import folium
from folium import Choropleth

# Crear mapa base centrado en Colombia
m = folium.Map(location=[4.5, -74], zoom_start=6)

# Crear capa coroplética de rendimiento promedio
folium.Choropleth(
    geo_data=gdf_muni.to_json(),
    data=gdf_muni,
    columns=["municipio", "rendimiento_prom"],
    key_on="feature.properties.mncp_nr",
    fill_color="YlGnBu",
    fill_opacity=0.7,
    line_opacity=0.3,
    legend_name="Rendimiento promedio (ton/ha)"
).add_to(m)

# Agregar control de capas
folium.LayerControl().add_to(m)

# Mostrar
m.save("C:/Users/paola/Analisis Geoespacial/Proyecto_01/Trabajo_Final/mapa_rendimiento_folium.html")
m
```

```python
import geopandas as gpd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import math

# ----------------------------------------
# Cargar datos
# ----------------------------------------
gdf = gpd.read_file("C:/Users/paola/Analisis Geoespacial/Proyecto_01/gdf_resultado_modelos_FINAL.shp")

# Seleccionar columnas de interés
cols = ["Dprtmnt", "mncp_nr", "Año", "Semestr", "Area", "Rend", "Rnd_Crt", "Bj_Rndm", "prcp_c_"]
df = gdf[cols].copy()

# Crear carpeta de salida si no existe
output_dir = "C:/Users/paola/Analisis Geoespacial/Proyecto_01/Trabajo_Final"
os.makedirs(output_dir, exist_ok=True)

# ----------------------------------------
# Diccionario de nombres largos para ejes y títulos
# ----------------------------------------
var_labels = {
    "Dprtmnt": "Departamento",
    "mncp_nr": "Municipio",
    "Año": "Año de observación",
    "Semestr": "Semestre",
    "Rnd_Crt": "Categoría de rendimiento",
    "Bj_Rndm": "Indicador de bajo rendimiento",
    "Area": "Área sembrada (ha)",
    "Rend": "Rendimiento (ton/ha)",
    "prcp_c_": "Precipitación acumulada (mm)"
}

# ----------------------------------------
# Gráficos de frecuencia para categóricas
# ----------------------------------------
cat_vars = ["Dprtmnt", "mncp_nr", "Año", "Semestr", "Rnd_Crt", "Bj_Rndm"]
n_cat = len(cat_vars)
cols = 3
rows = math.ceil(n_cat / cols)

sns.set(style="whitegrid")
fig_cat, axes_cat = plt.subplots(rows, cols, figsize=(18, 4 * rows))

for i, var in enumerate(cat_vars):
    ax = axes_cat.flatten()[i]
    sns.countplot(data=df, x=var, ax=ax, palette="Set2", order=df[var].value_counts().index)
    ax.set_title(var_labels.get(var, var), fontsize=12)
    ax.set_xlabel("")
    ax.set_ylabel("Frecuencia")
    ax.tick_params(axis='x', rotation=45)

# Borrar ejes vacíos si sobran
for j in range(i + 1, len(axes_cat.flatten())):
    fig_cat.delaxes(axes_cat.flatten()[j])

fig_cat.suptitle("📊 Frecuencias de variables categóricas", fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.96])
fig_cat.savefig(os.path.join(output_dir, "frecuencias_categoricas.png"), dpi=300)

# ----------------------------------------
# Histogramas + KDE para variables numéricas
# ----------------------------------------
num_vars = ["Area", "Rend", "prcp_c_"]
n_num = len(num_vars)

fig_num, axes_num = plt.subplots(1, n_num, figsize=(6 * n_num, 5))

for i, var in enumerate(num_vars):
    ax = axes_num[i]
    sns.histplot(df[var], kde=True, ax=ax, bins=30, color="cornflowerblue", edgecolor="black")
    ax.set_title(var_labels.get(var, var), fontsize=12)
    ax.set_xlabel(var_labels.get(var, var))
    ax.set_ylabel("Frecuencia")
    ax.grid(True, linestyle="--", alpha=0.5)

fig_num.suptitle("📈 Distribuciones de variables numéricas", fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.95])
fig_num.savefig(os.path.join(output_dir, "distribuciones_numericas.png"), dpi=300)
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import folium
from folium.plugins import HeatMap
import os

# --- Rutas ---
ruta_gpkg = r"C:\Users\paola\Analisis Geoespacial\Proyecto_01\Geopackage\Servicio-610\Municipios_Abril_2025.gpkg"
ruta_excel = r"C:\Users\paola\Analisis Geoespacial\Proyecto_01\Datos de cultivos\BaseDatosArroz_con_precipitacion_alta.xlsx"

# --- Leer capa 'Munpio' ---
gdf = gpd.read_file(ruta_gpkg, layer='Munpio')

# --- Función para normalizar texto ---
def normalizar(texto):
    if isinstance(texto, str):
        return texto.replace(" ", "").upper()
    return texto

# --- Normalizar nombres ---
gdf['municipio_norm_gpkg'] = gdf['MpNombre'].apply(normalizar)
gdf['departamento_norm'] = gdf['Depto'].apply(normalizar)

# --- Filtrar municipios de Meta y Casanare ---
gdf_filtrado = gdf[gdf['departamento_norm'].isin(['META', 'CASANARE'])]

# --- Proyectar a lat/lon y calcular centroides ---
gdf_latlon = gdf_filtrado.to_crs(epsg=4326)
gdf_latlon['centroide'] = gdf_latlon.geometry.centroid
gdf_latlon['lat'] = gdf_latlon['centroide'].y
gdf_latlon['lon'] = gdf_latlon['centroide'].x

# --- Leer Excel ---
df_excel = pd.read_excel(ruta_excel)
df_excel['municipio_norm_norm'] = df_excel['municipio_norm'].apply(normalizar)

# --- Merge con coordenadas desde el GPKG ---
df_merge = df_excel.merge(
    gdf_latlon[['municipio_norm_gpkg', 'lat', 'lon', 'geometry']],
    left_on='municipio_norm_norm',
    right_on='municipio_norm_gpkg',
    how='left'
)

# --- Limpiar columnas auxiliares ---
df_merge = df_merge.drop(columns=['municipio_norm_norm', 'municipio_norm_gpkg'])

# --- Crear mapa base centrado en Meta/Casanare ---
centro_mapa = [df_merge['lat'].mean(), df_merge['lon'].mean()]
m = folium.Map(location=centro_mapa, zoom_start=8)

# --- Agregar capa de calor ---
puntos = df_merge[['lat', 'lon']].dropna().values.tolist()
HeatMap(puntos, radius=15, blur=10, max_zoom=12).add_to(m)

# --- Crear carpeta si no existe y guardar ---
carpeta_salida = r"C:\Users\paola\Analisis Geoespacial\Proyecto_01\Trabajo_Final"
os.makedirs(carpeta_salida, exist_ok=True)
ruta_salida = os.path.join(carpeta_salida, "KDE_interactivo.html")

m.save(ruta_salida)
print(f"✅ Mapa de calor guardado en: {ruta_salida}")
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
from libpysal.weights import Queen
from esda.moran import Moran
import numpy as np

# 1. Usamos tu GeoDataFrame
gdf = gpd.read_file("C:/Users/paola/Analisis Geoespacial/Proyecto_01/gdf_resultado_modelos_FINAL.shp")

# 2. Crear matriz de pesos espaciales (adyacencia)
w = Queen.from_dataframe(gdf)
w.transform = 'r'

# 3. Calcular el valor rezagado espacial de 'Rend'
gdf['w_Rend'] = w.sparse.dot(gdf['Rend'])

# 4. Estandarizar
gdf['Rend_std'] = (gdf['Rend'] - gdf['Rend'].mean()) / gdf['Rend'].std()
gdf['w_Rend_std'] = (gdf['w_Rend'] - gdf['w_Rend'].mean()) / gdf['w_Rend'].std()

# 5. Graficar el diagrama de Moran
f, ax = plt.subplots(1, figsize=(6, 6))
sns.regplot(x='Rend_std', y='w_Rend_std', ci=None, data=gdf, line_kws={'color': 'r'})
ax.axvline(0, c='k', alpha=0.5)
ax.axhline(0, c='k', alpha=0.5)
ax.set_title('Moran Plot - Rendimiento agrícola', fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```

```python
from libpysal.weights import Queen
from esda.moran import Moran

# Matriz de pesos espaciales
w = Queen.from_dataframe(gdf, use_index=False)
w.transform = 'r'

# Cálculo del índice de Moran
mi = Moran(gdf['Rend'], w)

# Resultados limpios
print(f"Índice de Moran: {mi.I:.4f}")
print(f"p-valor (simulado): {mi.p_sim:.4f}")
print(f"z-score: {mi.z[0]:.2f}")  # <-- corregido aquí
```

```python
import numpy as np

gdf['Rend_log'] = np.log1p(gdf['Rend'])  # log(1 + Rend) evita log(0)
gdf['Rend_log_std'] = (gdf['Rend_log'] - gdf['Rend_log'].mean()) / gdf['Rend_log'].std()
gdf['w_Rend_log'] = w.sparse.dot(gdf['Rend_log'])
gdf['w_Rend_log_std'] = (gdf['w_Rend_log'] - gdf['w_Rend_log'].mean()) / gdf['w_Rend_log'].std()
```

```python
import numpy as np
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from mgwr.gwr import GWR
from mgwr.sel_bw import Sel_BW
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.axes_grid1 import make_axes_locatable

# -----------------------------
# 1. Cargar shapefile
# -----------------------------
gdf = gpd.read_file("C:/Users/paola/Analisis Geoespacial/Proyecto_01/gdf_resultado_modelos_FINAL.shp")

# -----------------------------
# 2. Preparar datos
# -----------------------------
# Asegúrate de que la proyección esté en metros (EPSG:3116 es Colombia Magna)
gdf = gdf.to_crs(epsg=3116)

# Log del rendimiento
gdf['Rend_log'] = np.log1p(gdf['Rend'])

# Variables predictoras
variables = ['prcp_c_', 'Area']

# Escalamiento de variables predictoras
scaler = StandardScaler()
X_std = scaler.fit_transform(gdf[variables])
y = gdf[['Rend_log']].values

# Coordenadas del centroide
coords = np.column_stack([gdf.geometry.centroid.x, gdf.geometry.centroid.y])

# -----------------------------
# 3. Seleccionar ancho de banda
# -----------------------------
bw = Sel_BW(coords, y, X_std).search()
print(f"Ancho de banda óptimo: {bw}")

# -----------------------------
# 4. Ajustar modelo GWR
# -----------------------------
gwr_model = GWR(coords, y, X_std, bw)
results = gwr_model.fit()

# -----------------------------
# 5. Guardar resultados en el GeoDataFrame
# -----------------------------
coef_names = ['Intercepto'] + variables
for i, name in enumerate(coef_names):
    gdf[f'gwr_{name}'] = results.params[:, i]

# R² local
gdf['local_R2'] = results.localR2

# -----------------------------
# 6. Graficar coeficientes
# -----------------------------
fig, axes = plt.subplots(1, len(variables), figsize=(16, 6))

for i, var in enumerate(variables):
    ax = axes[i]
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.1)

    gdf.plot(column=f'gwr_{var}',
             cmap='coolwarm',
             linewidth=0.1,
             edgecolor='white',
             legend=True,
             ax=ax,
             cax=cax)

    ax.set_title(f'Coeficiente local: {var}', fontsize=13)
    ax.set_axis_off()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
from mgwr.gwr import GWR
from mgwr.sel_bw import Sel_BW
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.axes_grid1 import make_axes_locatable
import os

# -----------------------------------
# 1. Cargar el shapefile
# -----------------------------------
gdf = gpd.read_file("C:/Users/paola/Analisis Geoespacial/Proyecto_01/gdf_resultado_modelos_FINAL.shp")

# -----------------------------------
# 2. Preprocesar datos
# -----------------------------------
# Asegurar que tiene CRS proyectado en metros
gdf = gdf.to_crs(epsg=3116)

# Variable dependiente: Rendimiento log-transformado
gdf['Rend_log'] = np.log1p(gdf['Rend'])

# Convertir 'Semestr' a dummy: I = 0, II = 1
gdf['Semestr_dummy'] = gdf['Semestr'].apply(lambda x: 1 if x == 'II' else 0)

# Variables independientes
variables = ['prcp_c_', 'Area', 'Semestr_dummy']
X_std = StandardScaler().fit_transform(gdf[variables])
y = gdf[['Rend_log']].values

# Coordenadas de los centroides
coords = np.column_stack([gdf.geometry.centroid.x, gdf.geometry.centroid.y])

# -----------------------------------
# 3. Seleccionar ancho de banda
# -----------------------------------
bw = Sel_BW(coords, y, X_std).search()
print(f"Ancho de banda óptimo: {bw}")

# -----------------------------------
# 4. Ajustar modelo GWR
# -----------------------------------
gwr_model = GWR(coords, y, X_std, bw)
results = gwr_model.fit()

# -----------------------------------
# 5. Guardar coeficientes locales
# -----------------------------------
coef_names = ['Intercepto'] + variables
for i, name in enumerate(coef_names):
    gdf[f'gwr_{name}'] = results.params[:, i]

# También guardar el R² local
gdf['local_R2'] = results.localR2

# -----------------------------------
# 6. Guardar gráficas de coeficientes locales
# -----------------------------------
# Crear carpeta de salida si no existe
ruta_salida = "C:/Users/paola/Analisis Geoespacial/Proyecto_01/GWR_Maps/"
os.makedirs(ruta_salida, exist_ok=True)

for var in variables:
    col = f'gwr_{var}'
    
    fig, ax = plt.subplots(1, figsize=(8, 6))
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.1)

    gdf.plot(column=col,
             cmap='coolwarm',
             linewidth=0.3,
             edgecolor='white',
             legend=True,
             ax=ax,
             cax=cax)

    ax.set_title(f'Coeficiente local: {var}', fontsize=14)
    ax.set_axis_off()
    plt.tight_layout()
    
    # Guardar imagen
    nombre_archivo = f"coeficiente_local_{var}.png"
    plt.savefig(os.path.join(ruta_salida, nombre_archivo), dpi=300)
    plt.close()

# Guardar gráfico del local R2
fig, ax = plt.subplots(1, figsize=(8, 6))
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)

gdf.plot(column='local_R2',
         cmap='YlGn',
         linewidth=0.3,
         edgecolor='white',
         legend=True,
         ax=ax,
         cax=cax)

ax.set_title('R² local (ajuste del modelo)', fontsize=14)
ax.set_axis_off()
plt.tight_layout()
plt.savefig(os.path.join(ruta_salida, "local_R2.png"), dpi=300)
plt.close()

# -----------------------------------
# 7. Imprimir resumen estadístico
# -----------------------------------
resumen = gdf[[f'gwr_{v}' for v in variables] + ['local_R2']].describe().T
print(resumen[['min', 'mean', 'max']])
```

|------------------------------------|----------------------------------------|----------------|
| **Intercepto (media)**             | 0.351                                  | Valor medio del rendimiento esperado en ausencia de efectos espaciales. Es estable, con una desviación estándar baja (≈0.01), lo que indica buena precisión. |
| **Precisión observacional**        | 15.34 (equivale a varianza ≈ 0.065)    | Baja varianza residual → el modelo explica bien los datos. |
| **Precisión del efecto espacial**  | 0.531 (varianza ≈ 1.88)                | Existe heterogeneidad espacial importante: algunas regiones tienden a sobre- o sub-rendir en comparación con el promedio. |
| **DIC (Deviance Information Criterion)** | 229.35                            | Bajo, lo que indica buen ajuste general del modelo. Si se compara con un modelo sin efecto espacial, este valor sería claramente mejor. |

---

### 🗺️ ¿Qué indica el modelo?

- El rendimiento promedio estimado es **0.351**, con intervalos estrechos, lo cual brinda confianza en la predicción.
- La varianza espacial (1 / precisión ≈ 1.88) indica que **sí hay diferencias regionales significativas** en el rendimiento, detectadas por el modelo.
- El **DIC bajo (229.35)** indica que el modelo tiene un buen equilibrio entre ajuste y complejidad.
- Este modelo permite construir **mapas de predicción y efectos espaciales** que aportan información útil para la toma de decisiones territoriales y agrícolas.

## 🧠 Conclusiones Generales del Análisis Espacial del Rendimiento Agrícola

Este estudio aplicó tres enfoques estadísticos espaciales distintos para analizar la distribución del rendimiento agrícola en función de variables como la precipitación acumulada, el área sembrada y el semestre de cultivo. A continuación, se presentan las principales conclusiones derivadas de los modelos ICAR Binomial, ICAR Gaussiano y GWR.

---

### 🔹 1. Modelo ICAR Binomial

- **Objetivo**: Modelar la probabilidad de que un municipio tenga un rendimiento alto (1) o bajo (0).
- **Resultado principal**: Las probabilidades predichas se concentraron entre 0.49 y 0.51 para todos los municipios.
- **DIC (criterio de ajuste)**: 908, lo que indica un **ajuste deficiente**.
- **Efecto espacial**: Prácticamente nulo.

📉 **Conclusión**:  
El modelo binomial **no logra capturar la variabilidad real** del rendimiento. La binarización de la variable original provoca pérdida de información y no permite aprovechar la riqueza de los datos continuos.

---

### 🔸 2. Modelo ICAR Gaussiano

- **Objetivo**: Modelar directamente el rendimiento agrícola como variable continua.
- **Intercepto estimado**: 0.351
- **Varianza espacial significativa**: precisión ≈ 0.53 → se evidencia **estructura espacial en los rendimientos**.
- **DIC**: 229 → **excelente ajuste** comparado con el binomial.

📊 **Conclusión**:  
Este modelo permite identificar patrones espaciales en el rendimiento. Es más realista y útil para generar **mapas de predicción y priorización territorial**. Representa mejor la heterogeneidad geográfica del fenómeno agrícola.

---

### 🌍 3. Modelo GWR (Geographically Weighted Regression)

- **Objetivo**: Explorar cómo varían localmente los efectos de las variables explicativas sobre el rendimiento.
- **Coeficientes locales**:
  - `prcp_c_`: leve efecto negativo promedio, pero con **alta variabilidad local**.
  - `Area`: efecto positivo generalizado.
  - `Semestr_dummy`: efecto negativo leve (semestre II).
- **R² local**: promedio = 0.78, mediana = 0.83 → **muy buen ajuste local**.
- **Bandwidth**: 53 municipios → vecindad espacial utilizada para el suavizado.

🧭 **Conclusión**:  
El modelo GWR **destaca zonas con relaciones diferenciales** entre clima, área y rendimiento. Permite un análisis fino y ajustado a la realidad de cada territorio, apoyando decisiones localizadas y efectivas.

---

### ✅ Conclusión Integrada

| Modelo         | Tipo de variable | Ajuste (DIC / R²) | Captura patrón espacial | Utilidad práctica |
|----------------|------------------|--------------------|--------------------------|--------------------|
| ICAR Binomial  | Binaria          | DIC ≈ 908          | No                    | Limitada           |
| ICAR Gaussiano | Continua         | DIC ≈ 229          | Sí                    | Alta               |
| GWR            | Continua + Local | R² medio ≈ 0.78    | Sí (localizado)       | Muy alta           |

📌 **Recomendación**:  
- Usar el **modelo ICAR Gaussiano** para modelado general de rendimiento con mapas de predicción confiables.  
- Complementar con **GWR** para análisis de sensibilidad local y para diseñar políticas agrícolas diferenciadas por zona.  
- Evitar el modelo binomial en fenómenos donde la continuidad de la variable tiene relevancia explicativa.

---

### 🔚 Reflexión Final

La modelación espacial permite evidenciar que **el rendimiento agrícola no es uniforme** y depende de múltiples factores que **varían regionalmente**. Aplicar modelos avanzados como ICAR y GWR proporciona herramientas robustas para apoyar la **planificación agrícola sostenible**, la **optimización de recursos** y la **adaptación al cambio climático** con enfoque territorial.


```python
import geopandas as gpd
import matplotlib.pyplot as plt
from libpysal.weights import Queen
from libpysal import examples
from shapely.geometry import LineString

# 1. Cargar tu shapefile
gdf = gpd.read_file("C:/Users/paola/Analisis Geoespacial/Proyecto_01/resultados_finales.gpkg")
gdf = gdf.to_crs(epsg=3116)  # Asegúrate de tener CRS proyectado

# 2. Crear matriz de vecinos tipo Queen
w_queen = Queen.from_dataframe(gdf)

# 3. Calcular centroides para conectar los vecinos
centroids = gdf.geometry.centroid

# 4. Crear líneas entre municipios vecinos
lines = []
for i, neighbors in w_queen.neighbors.items():
    for j in neighbors:
        line = LineString([centroids[i], centroids[j]])
        lines.append(line)

# 5. Crear GeoDataFrame con las líneas
lines_gdf = gpd.GeoDataFrame(geometry=lines, crs=gdf.crs)

# 6. Graficar
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, facecolor='lightgrey', edgecolor='white')
lines_gdf.plot(ax=ax, color='red', linewidth=0.8)

ax.set_title("Relaciones espaciales tipo Queen entre municipios", fontsize=14)
ax.set_axis_off()
plt.tight_layout()
plt.savefig("matriz_queen_vecinos.png", dpi=300)
plt.show()
```
